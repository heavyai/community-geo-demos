```python
import geopandas as gpd
```

```python
dateparse = lambda dates: [pd.datetime.strptime(d, '%Y-%m-%d') for d in dates]
infile = 'HistoricalFuelTreatments/USFS_Hazard_Fuel_Teatmentt_Activity_webmerc.shp'
treatments = gpd.read_file(infile) #, parse_dates=['REV_DATE'], date_parser=dateparse)
```

```python
treatments.head(1)
```

```python
treatments.columns
```

```python
treatments.REV_DATE.unique()
```

```python
#treatments['tdate'] = treatments.REV_DATE.str.split('-')[0]
treatments['date'] = treatments.REV_DATE.str.split('-')
```

```python
treatments['date'].sort_values().head()
```

```python
treatments['date'].dropna(inplace=True)
```

```python
treatments['date'] = treatments['date'].apply(lambda x: x[0])
```

```python
treatments.head()
```

```python
treatments.pivot_table(index='ACTIVITY', 
               columns='date',
               aggfunc='size',
               fill_value=0)
```

```python
treatments.pivot_table(index='ACTIVITY', 
               columns='date',
               aggfunc='size',
               fill_value=0).sum()
```

```python
treatments.pivot_table(index='ACTIVITY', 
               columns='date',
               aggfunc='size',
               fill_value=0).sum().median()
```
